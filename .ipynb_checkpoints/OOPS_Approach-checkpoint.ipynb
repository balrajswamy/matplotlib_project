```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

docstring = """plt.subplots() in Matplotlib is a versatile function for creating a grid of subplots,\nwhich allows you to display multiple plots in one figure.\n It’s particularly useful when you need to compare different visualizations side by side or in a grid layout.                """
print(docstring)

x = [1,2,3,4,5,6]
y = [23,35,27,20,18,45]

fig, ax = plt.subplots()
ax.set_title("Graph is drawn using subplots")

ax.plot(x,y)
plt.show()
```

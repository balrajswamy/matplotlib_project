```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
import numpy as np

config_dir = mpl.get_configdir()
print("Config Directory:", config_dir)
cache_dir = mpl.get_cachedir()
print("Cache Directory:", cache_dir)
print()

x = [1,2,3,4,5,6]
y = [23,35,27,20,18,45]

plt.plot(x, y, color='blue', linestyle='-', linewidth=2, marker='o', markersize=6, label="Line chart")
plt.title("Subscribers Count")
plt.xlabel("Date")
plt.ylabel("No of Persons")
plt.legend() 

plt.show()
```
